```python
s = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],

    
    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],
    
    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
])
```

```python
np.sum(s , axis = 0)
```

```python
np.sum(s,axis=1)
```

```python
z = np.sum(s,axis=0)
m = np.sum(s,axis=1)

if np.all(z == 45) and np.all(m == 45):
    print("all rows and columns have sum 45")
else:
    print("Invalid Sudoku")
```

```python
valid = True

for row in s:
    for element in row:
        if not(1<=element <=9):
            valid = False

if valid:
    print("all numbers are in bw 1 and 9")
else:
    print("numbers are upper than 9")
```

```python
np.sum(s,axis = 0) # 0 for columns
```

```python
b = np.sum(s,axis = 1)
b == 45
```

```python
c = np.sum(s,axis=1)   # this was for rows
for i in c:
    if i != 45:
        print("sudoku is not valid")
        break
else:
    print("for rows it is valid")
```

```python
d = np.sum(s,axis=0)    # for columns
for i in d:
    if i != 45:
        print("sudoko is not valid")
        break
else:
    print("for columns it is valid")
```

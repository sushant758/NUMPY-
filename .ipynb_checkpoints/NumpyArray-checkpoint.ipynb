## Spliting Array

```python
c = np.arange(16).reshape(4,4)
c
```

```python
np.hsplit(c,2)   # this can split it into two parts , and 2 means in two parts
```

```python
np.hsplit(c,2)   
```

```python
# u cant divide it into 3 parts , bcz it has not equal sitribution for 3
np.hsplit(c,3)
```

```python
np.hsplit(c,4)
```

```python
# vertical split
a = np.vsplit(c,4)
a
```

```python
for i in a:
    print(i)
    
```

```python
# it get splitted into 2d matrix
```

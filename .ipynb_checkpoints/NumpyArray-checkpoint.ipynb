# Array Generation Function

```python
arr = np.arange(1,11)
arr
```

```python
arr = np.arange(1,11,2)
arr
```

```python
# generate some zeros
arr = np.zeros(6)
arr
```

```python
# generate zeros as a matrix
arr = np.zeros((7,4))
arr
```

```python
arr = np.ones((8,4))
arr
```

```python
arr = np.linspace(1,20,4)  # equally space distribution
arr
```

```python
arr = np.linspace(0,1,100)  #it means between 0 and 1 i need 100 elements , and bw them 
arr                          # space are equally distributed
```

# Random Generation function

```python
# random number generation

# we will use random library

np.random.rand(20)
```

```python
# note = its a vector bcz it open and closes bw one bracket = [] , 
# and if its [[]] then it is a matrix
```

```python
# when a value comes bw 0 and 1 its called normalization
# and when a value comes bw -3 and 3 its called standardization
```

```python
np.random.randn(20)    # its standardization
```

```python
# generate random numbers or integera
np.random.randint(20)   # generated random no between 0 and 20
```

```python
np.random.randint(5,50,10) # 10 random no generated bw 5 and 50
```

# Array Methods

```python
arr
```

```python
# to find minimum elements
arr.min()
```

```python
arr.max()
```

```python
# sum , sum of all elements
arr.sum()
```

```python
# sum of what i want , any row or column
np.sum(arr,axis = 0)  # horizontal called 1
```

```python
np.sum(arr,axis=1)
```

```python
# find mean  , means middle number
arr.mean()
```

```python
# standard deviation
arr.std()
```

```python
# maximum element index 
arr.argmax()
```

```python
arr.argmin()
```
